```python
class Persona:
    nombre: str
    edad: int
    ciudad: str
    
    def __init__(self, nombre: str, edad: int, ciudad: str):
        self.nombre = nombre
        self.edad = edad
        self.ciudad = ciudad

    def __str__(self ):
      return f"Persona : nombre: {self.nombre}, edad: {self.edad}, ciudad: {self.ciudad}"
    

    def presentarse(self):
        print(f"Hola, mi nombre es {self.nombre}, y mi edad {self.edad}")
    
    def cumplir_años(self):
        self.edad+=1

    def cambiar_ciudad(self, nueva_ciudad):
        self.ciudad=nueva_ciudad

    def get_edad(self):
        return self.edad

    def set_edad(self, nueva_edad):
        if nueva_edad >= 0:
                self.edad = nueva_edad
        else:
            print("La edad no puede ser un numero negativo.")

    

# Crear una instancia de la clase Persona
persona1 = Persona("Carlos", 35, "Chiclayo")
persona2 = Persona("Matias", 15, "Tacna")
persona3 = Persona("Jose", 25, "Lima")

print("----EJERCIO 1-----")
print(persona1)
persona1.edad= 32
print(persona1)

print(persona2)
persona2.ciudad= "Lima"
print(persona2)

print(persona3)
persona3.ciudad="Arequipa"
print(persona3)


print("----EJERCIO 2-----")
persona1.presentarse()
persona1.cumplir_años()
print(persona1)


print("----EJERCIO 3-----")
class Estudiante(Persona):
    grado:int

    def __init__(self, nombre, edad, ciudad, grado):
        super().__init__(nombre, edad, ciudad)
        self.grado = grado

    def estudiar(self):
        print(f" {self.nombre} esta estudiando")



estudiante = Estudiante("Jorge", 15, "Talara", 4)
estudiante.presentarse()
estudiante.cumplir_años()
estudiante.cambiar_ciudad("Iquitos")
print(estudiante)
estudiante.estudiar()


print("----EJERCIO 4-----")

class Animal():
    def hacer_sonido(self):
        print("Mi sonido es generico")



class Perro(Animal):
    def hacer_sonido(self):
       print("GUAU GUAU")

class Gato(Animal):
    def hacer_sonido(self):
       print("MIAU MIAU")



#----funcion HACER_QUE_SUENEN----
def hacer_que_suenen(lista_animales):
    for animal in lista_animales:
        animal.hacer_sonido()


perro = Perro()
gato = Gato()

lista_animales =[perro, gato]
hacer_que_suenen(lista_animales)


print("----EJERCIO 5-----")
persona4 = Persona("Eduardo", 28, "Lima")
print(persona4.get_edad())  
persona4.set_edad(35)
print(persona4.get_edad()) 
persona4.set_edad(-5) 

```
